```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

bunch = load_boston()
target = bunch.target 

df = pd.DataFrame(data=bunch.data, columns=bunch.feature_names)
df.head()
```

```python
df.hist(column=bunch.feature_names[0])
df.hist(column=bunch.feature_names[1])
```

```python
scores = []
for column_name in df.columns[0:2]:
    column = df[column_name].values.reshape(-1, 1)
    reg = LinearRegression().fit(column, target)
    predicted_target = reg.predict(column)
    scores.append((column_name, reg.score(column, target)))

    plt.scatter(column, target, label="real")
    plt.scatter(column, predicted_target, label="predicted")
    plt.title(f'Linear model score ={score:.3f} for collumn {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Median of owner-occupied homes in $1000')
    plt.legend()
    plt.figure()
print(scores)
```
